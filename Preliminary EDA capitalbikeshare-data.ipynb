```python
import plotly.graph_objects as go

# Exclude date columns
columns_to_exclude = ['Start date', 'End date', 'Duration']

unique_values = {
    '2018 January': bikedata_2018_01.nunique().drop(labels=columns_to_exclude),
    '2017 Q1': bikedata_2017_q1.nunique().drop(labels=columns_to_exclude),
    '2017 Q2': bikedata_2017_q2.nunique().drop(labels=columns_to_exclude),
    '2017 Q3': bikedata_2017_q3.nunique().drop(labels=columns_to_exclude),
    '2017 Q4': bikedata_2017_q4.nunique().drop(labels=columns_to_exclude)
}

datasets = ['2018 January', '2017 Q1', '2017 Q2', '2017 Q3', '2017 Q4']

# Creating the bar chart
fig = go.Figure()
for dataset in datasets:
    fig.add_trace(go.Bar(x=unique_values[dataset].index, y=unique_values[dataset].values, name=dataset))

fig.update_layout(
    title='Unique Values in Each Column by Dataset (Excluding Dates)',
    xaxis_title='Columns',
    yaxis_title='Number of Unique Values',
    barmode='group'
)
fig.show()

```
